```python
from sklearn.datasets import load_breast_cancer
```

```python
#binary classifier to predict cancer using parameters
cancer =load_breast_cancer()
print(cancer)
```

```python
import numpy as np
import pandas as pd
cancer_df = pd.DataFrame(data=cancer.data,
                           columns=cancer.feature_names)
```

```python
cancer_df['target']=cancer.target
print(cancer_df.head())
print(cancer_df.tail())
print(cancer_df.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
print(cancer_df.corr())
```

```python
y=cancer_df['target']
X= cancer_df.drop('target',axis=1)
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.4,random_state=47)
tree = DecisionTreeClassifier()
tree.fit(X_train,Y_train)
```

```python
tree.predict(X_test)
```

```python
print(Y_test)
```

```python
print("Accuracy on test set: {:.3f}".format(tree.score(X_test,Y_test)))
```

```python
print("Accuracy on train set: {:.3f}".format(tree.score(X_train,Y_train)))
```

```python
tree1 = DecisionTreeClassifier(max_depth= 7)
tree1.fit(X_train,Y_train)
```

```python
print("Accuracy on test set: {:.3f}".format(tree1.score(X_test,Y_test)))
```

```python
tree2 = DecisionTreeClassifier(max_depth= 12)
tree2.fit(X_train,Y_train)
```

```python
print("Accuracy on test set: {:.3f}".format(tree2.score(X_test,Y_test)))
```

```python
# Here varying the depth of the node can help us achieve higher result and accuracy on testing data and avoid overfitting
# max_depth of 6-7 is ideal and beyond that the accuracy begins to decline. 
```

```python
import matplotlib.pyplot as plt
cancer_df['target'].value_counts().plot.bar(color = ["r","b"]) 
plt.title('Cancer prediction on dataset') 
plt.show()
```

```python
predictions = tree.predict(X_test)
predictions_df = pd.DataFrame(predictions, columns=['Predicted'])
counts = predictions_df['Predicted'].value_counts()
counts.plot.bar(color=['r','b'])
plt.title("cancer prediction based on trained model")
```

```python
# matplotlib was used to visualise difference in original dataset and trained model
# This can be compared with a logistic regression algorithm
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(solver="liblinear",max_iter=500)
model.fit(X_train,Y_train)
print("Accuracy score on  testing data : {:3f}".format(model.score(X_test,Y_test)))
print("Accuracy score on training data: {:3f}".format(model.score(X_train,Y_train)))
```
